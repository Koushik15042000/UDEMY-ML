```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Graphs can be plotted using 2 ways,
# 1. Functional Method
# 2. Object Oriented Method
```

```python
# Functional Method
plt.plot(x,y)
```

```python
# Add X & Y lables and Title of the Plot
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('DEMO GRAPH')
```

```python
# Plot 2 Graphs on the same Canvas
# subplot(no_of_rows, no_f_cols, plot_no)
plt.subplot(1,2,1)
# 'r' stands for red Color
plt.plot(x,y,'r')
plt.xlabel('X Label 1')
plt.ylabel('Y Label 1')
plt.title('PLOT 1')

plt.subplot(1,2,2)
# 'r' stands for red Color
plt.plot(x,y,'b')
plt.xlabel('X Label 2')
plt.ylabel('Y Label 2')
plt.title('PLOT 2')
```

```python
# Object Oriented Method
#fig.axes(length, bottom , width, height)
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y,'r')
axes1.set_title('LARGER PLOT')

axes2 = fig.add_axes([0.2,0.4,0.3,0.4])
axes2.set_title('SMALLER PLOT')
axes2.plot(y,x,'b')
```
