## Plot Appearance

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y)
# Set Colors of the curve
ax.plot(x,y,color = 'orange')
# We can Use RGB Hex Codes in color
#ax.plot(x,y,color = '#FF8C00')
#ax.plot(x,y,color = 'green')
#ax.plot(x,y,'r')
```

```python
# Default Line Width is 1
# Note: bOth linewidth or lw as parameters will work
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color = 'orange', linewidth = 5)
ax.plot(x,y,color = 'orange', lw = 5)
```

```python
# Control Transparency of the curve using Line Visibility - alpha parameter
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'orange', lw = 5, alpha = 0.5)
```

```python
lineStyles = {'-': '_draw_solid', '--': '_draw_dashed', '-.': '_draw_dash_dot', 
              ':': '_draw_dotted', 'None': '_draw_nothing', ' ': '_draw_nothing', '': '_draw_nothing'}
lineStyles_array = np.array(lineStyles)
```

```python
lineStyles_array
```

```python
# Control Style of the curve using Line Style - linestyle parameter
# Note: bOth linestyle or ls as parameters will work
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# '--' stands for Dashed Line
# '-.' stands for Dashed Dot Line
#ax.plot(x,y,color = 'orange', lw = 5, linestyle = '-.')
ax.plot(x,y,color = 'orange', lw = 5, ls = '-.')
#ax.plot(x,y,color = 'orange', lw = 5, linestyle = 'steps')
#ax.plot(x,y,color = 'orange', lw = 5, alpha = 0.5, linestyle = '--')
```

```python
# To Mark each point on Curve, we make use use marker as parameter and giving a string as input, most common string is 'o'
# We can change marker Size with the parameter markersize
# We can add color to Marker using markerfacecolor
# We can add Marker Edge Width using markeredgewidth
# We can add Marker Edge Color using markeredgecolor
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color = 'orange', lw = 3, ls ='-', marker = 'o')
#ax.plot(x,y,color = 'orange', lw = 3, ls ='-', marker = 'x')
#ax.plot(x,y,color = 'orange', lw = 3, ls ='-', marker = '1')
ax.plot(x,y,color = 'purple', lw = 1, ls ='-', marker = 'o', markersize = 20, 
        markerfacecolor = 'yellow', markeredgewidth = 3, markeredgecolor = 'black')
fig.savefig('custom_plot.png', dpi = 100)
```

## Control over Axis Appearance

Plot Range

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, color = 'purple', lw = 2, ls = '--')
# If we wanna show plot only between a range for x-axis and y-axis
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

## Special Plot Types

There are many specialized plots that we can create, such as bar plots, histograms, scatter plots and much more. Most of these plots we can plot using seaborn - a statistical plotting library for Python. A few examples of these plots are:

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000), 100)
plt.hist(data)
```

```python
data = [np.random.normal(0,std, 100) for std in range(1,4)]
# rectangular Box Plot
plt.boxplot(data, vert = True, patch_artist = True)
```
