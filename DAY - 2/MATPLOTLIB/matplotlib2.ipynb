```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 2)
#axes, this is an iterable with 2 MatplotLib Axes objects
axes[0].plot(x,y)
axes[1].plot(y,x)
#for current_ax in axes:
#    current_ax.plot(x,y)
# tight_layout() method fixes issues of overlapping
plt.tight_layout()
```

fig, axes = plt.subplots(nrows = 3,ncols = 3)

## Figure size and DPI

```python
# Here DPI is Density per independent Inhces
#fig = plt.figure(figsize = (3,2), dpi = 100)
#fig = plt.figure(figsize = (3,2))
fig = plt.figure(figsize = (8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```
